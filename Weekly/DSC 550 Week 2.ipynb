```python
df['txt'] = df['txt'].str.lower()
```

```python
df['txt']= df['txt'].str.replace('.','')
df['txt']= df['txt'].str.replace(',','')
df['txt']= df['txt'].str.replace('!','')
df['txt']= df['txt'].str.replace("'",'')
df['txt']= df['txt'].str.replace('''[''','')
df['txt']= df['txt'].str.replace(']','')
df['txt']= df['txt'].str.replace('?','')
df
```
